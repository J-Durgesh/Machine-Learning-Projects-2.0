```python
from sklearn.preprocessing import LabelEncoder
encd=LabelEncoder()
```

```python
df.drop(columns=['car_ID','CarName'],axis=1,inplace=True)
```

```python
# Split data into X and Y 

x=df[[ 'symboling', 'fueltype', 'aspiration', 'doornumber',
       'carbody', 'drivewheel', 'enginelocation', 'wheelbase', 'carlength',
       'carwidth', 'carheight', 'curbweight', 'enginetype', 'cylindernumber',
       'enginesize', 'fuelsystem', 'boreratio', 'stroke', 'compressionratio',
       'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42,shuffle=True)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

###  Method 4

```python
from sklearn.feature_selection import SelectKBest,GenericUnivariateSelect,f_regression,chi2

k_best_selector = SelectKBest(score_func=f_regression, k=2)
X_train_selected_k = k_best_selector.fit_transform(X_train, y_train)
X_test_selected_k = k_best_selector.transform(X_test)
```

```python
k_test = pd.DataFrame({'Feature': (X_train.columns), 
                       'F_Score': k_best_selector.scores_, 
                       'P_Value': k_best_selector.pvalues_})

k_test=k_test.sort_values(ascending=False,by='F_Score')
```

```python
GenericUnivariate_selector = GenericUnivariateSelect(score_func=f_regression ,mode='k_best', param=3)
X_train_selected = GenericUnivariate_selector.fit_transform(X_train, y_train)
X_test_selected = GenericUnivariate_selector.transform(X_test)
```

```python
g_test = pd.DataFrame({
    'Feature': X_train.columns,
    'Score': GenericUnivariate_selector.scores_,
    'Selected': GenericUnivariate_selector.get_support()
})

g_test=g_test.sort_values(ascending=False,by='Score')
```

# Model

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from xgboost import XGBRegressor

xgb=XGBRegressor(objective='reg:squarederror')
```

```python
# From GenericUnivariateSelect

xgb.fit(X_train_selected,y_train)
score=xgb.score(X_test_selected,y_test)
y_predict_G=xgb.predict(X_test_selected)
print('Score:',score)
print('mae Score:',mean_absolute_error(y_predict_G,y_test))
print('mse Score:',mean_squared_error(y_predict_G,y_test))
print('r2 Score:',r2_score(y_predict_G,y_test))
```

```python
# From SelectKBest

xgb.fit(X_train_selected_k,y_train)
score=xgb.score(X_test_selected_k,y_test)
y_predict_k=xgb.predict(X_test_selected_k)
print('Score:',score)
print('mae Score:',mean_absolute_error(y_predict_k,y_test))
print('mse Score:',mean_squared_error(y_predict_k,y_test))
print('r2 Score:',r2_score(y_predict_k,y_test))
```
